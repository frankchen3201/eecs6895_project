```python
##extract caption files for youtube videos
import json
import os
import re
import time

f=open('train_val_videodatainfo.json')
data=json.load(f)
videos=data['videos']
for video in videos:
    cur=dict(video)
    vid=cur['video_id']
    url=cur['url']
    print(vid)
    
    p=re.compile("v\=(.+)")
    videoid=p.findall(url)[0]
    
    time.sleep(0.05)
    os.system("python>temp.txt d:/deep/caption.py --videoid=\""+videoid+"\" --action=\"list\"")

    temp=open('temp.txt')
    text=temp.readlines()
    temp.close()

    p=re.compile("\((.+)\)")

    captionid=p.findall(text[0])[0]

    time.sleep(0.05)
    os.system("python>temp.txt d:/deep/caption.py --captionid=\""+captionid+"\" --language=\"en\" --action=\"download\"")
    
    temp=open('temp.txt')
    text=temp.readlines()
    temp.close()
    print(text[0])

    out=open('d:/deep/caption/'+vid+'.word','w')
    for i in range(2,len(text)):
        if (text[i][0]=='\n'):
            i+=2
        out.write(text[i])
    out.close()
    
f.close()   
```
