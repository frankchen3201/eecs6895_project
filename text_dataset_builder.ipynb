```python
##coco
import json
f=open('D:/deep/annotations/instances_val2014.json')
data=json.load(f)
data.keys()
```

```python
data['categories']
```

```python
##count categories for images
chck={}
ast=set()
cate={}
for i in range(0,len(data['annotations'])):
    m=data['annotations'][i]
    r=m['image_id']
    y=m['category_id']
    if r in ast:
        chck[r]+=1
    else:
        chck[r]=1
        ast.add(r)
        cate[r]=y
```

```python
##select images with only 1 
data_s={}
cl=set()
for x in ast:
    if chck[x]==1:
        cl.add(x)
        data_s[x]=cate[x]
len(data_s)
```

```python
test_id=list(cl)
s_id=sorted(test_id)
fout=open('D:/deep/annotations/out_id.txt','w')
for li in s_id:
    fout.write(str(li)+"\n")
fout.close()
```
