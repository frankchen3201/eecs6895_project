```python
f=open("d:/deep/annotations/out_id.txt")
text=f.readlines()
ids=set()
for line in text:
    ids.add(int(line))
f.close()
```

```python
import urllib 
import urllib2
url="http://mscoco.org/images/"
for xid in ids:
    print ("downloading:"+str(xid))
    urllib.urlretrieve(url+str(xid),"d:/deep/images/"+str(xid)+".jpg")
```
